# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import plotly.express as px
```

# Import Dataset

```python
df= pd.read_csv('/content/Education_Data.csv')
```

```python
df.head()
```

# Data Cleaning

```python
# check columns name and white spaces

df.columns
```

```python
# change column names by best practice situation

df.columns = ['gender', 'nationality', 'place_of_birth', 'stage_id', 'grade_id',
       'section_id', 'topic', 'semester', 'relation', 'raised_hands',
       'visited_resources', 'announcements_view', 'discussion',
       'parent_answering_survey', 'parent_school_satisfaction',
       'student_absence_days', 'class']
```

```python
df.columns
```

```python
# check data type

print(df.dtypes)
```

```python
# check null values

print(df.isnull().sum())
```

```python
# check duplicated values

print(f'sum of duplicated values= {df.duplicated().sum()}')
```

```python
# drop duplicated values

df.drop_duplicates(inplace=True)
```

```python
# check duplicated values again

print(f'sum of duplicated values= {df.duplicated().sum()}')
```

```python
# print every unique value for each object column in a for loop start is 0, end is len() of object column list, step increase by 1

for col in df.columns:
    if (df[col].dtypes == 'object') & (len(df[col].unique()) <= 10):
        print(f"column name {col}:")
        print(f"{df[col].unique()}\n")
```

```python
# change data by best practice situation

for x in df.index :
    if df.loc[x, 'nationality'] == 'KW' :
        df.loc[x, 'nationality'] = 'Kuwait'
    elif df.loc[x, 'nationality'] == 'lebanon' :
        df.loc[x, 'nationality'] = 'Lebanon'

for x in df.index :
    if df.loc[x, 'place_of_birth'] == 'KuwaIT' :
        df.loc[x, 'place_of_birth'] = 'Kuwait'
    elif df.loc[x, 'place_of_birth'] == 'lebanon' :
        df.loc[x, 'place_of_birth'] = 'Lebanon'

for x in df.index :
    if df.loc[x, 'gender'] == 'M' :
        df.loc[x, 'gender'] = 'Male'
    elif df.loc[x, 'gender'] == 'F' :
        df.loc[x, 'gender'] = 'Female'

for x in df.index :
    if df.loc[x, 'semester'] == 'F' :
        df.loc[x, 'semester'] = 'First'
    elif df.loc[x, 'semester'] == 'S' :
        df.loc[x, 'semester'] = 'Second'

for x in df.index :
    if df.loc[x, 'class'] == 'M' :
        df.loc[x, 'class'] = 'Middle'
    elif df.loc[x, 'class'] == 'L' :
        df.loc[x, 'class'] = 'Low'
    elif df.loc[x, 'class'] == 'H' :
        df.loc[x, 'class'] = 'High'
```

```python
# check again and print every unique value for each object column in a for loop start is 0, end is len() of object column list, step increase by 1

for col in df.columns:
    if (df[col].dtypes == 'object') & (len(df[col].unique()) <= 10):
        print(f"column name {col}:")
        print(f"{df[col].unique()}\n")
```

```python
# check outliers

df_num= df.select_dtypes(exclude= ['object', 'bool', 'datetime64[ns]'])

print(df_num.describe().round(2))
```

```python
for col in df_num.columns:
    plt.figure(figsize= (10, 5))
    sns.boxplot(data= df_num,
                x= col)
    plt.title(f'discribe values of {col}')
    plt.show()
```

# Data Analysis

```python
sns.heatmap(data= df.corr(numeric_only= True), square= True, cmap= 'coolwarm', vmax= 1, vmin= -1, annot= True)
plt.title('correlation matrix')
plt.show()
```

# Relations  
**good relation between visited_resources & raised_hands 0.69**  
**good relation between announcements_view & raised_hands 0.64**  
**good relation between visited_resources & announcements_view 0.59**


```python
# Relation Between Num of Times Student Visted Resources & Num of Times Student Raised His Hand

plt.figure(figsize= (25, 5))

plt.subplot(1, 3, 1)
sns.regplot(data= df, x= 'visited_resources', y= 'raised_hands')
plt.title('Relation Between visited_resources & raised_hands')

plt.subplot(1, 3, 2)
sns.histplot(data= df, x= 'visited_resources', bins= 10)
plt.title('Distribution of visited_resources')

plt.subplot(1, 3, 3)
sns.histplot(data= df, x= 'raised_hands', bins= 10)
plt.title('Distribution of raised_hands')

plt.tight_layout()
plt.show()
```

# Insights  
**max visited resources times between 80 to 89 times**  
**max raised hands times between 10 to 19 times**  
**every time visted resources times increased the raised hands increased**

```python
# Relation Between announcements_view & raised_hands

plt.figure(figsize= (25, 5))

plt.subplot(1, 3, 1)
sns.regplot(data= df, x= 'announcements_view', y= 'raised_hands')
plt.title('Relation Between announcements_view & raised_hands')

plt.subplot(1, 3, 2)
sns.histplot(data= df, x= 'announcements_view', bins= 10)
plt.title('Distribution of announcements_view')

plt.subplot(1, 3, 3)
sns.histplot(data= df, x= 'raised_hands', bins= 10)
plt.title('Distribution of raised_hands')

plt.tight_layout()
plt.show()
```

# Insights  
**max announcements_view times between 10 to 20 times**  
**max raised hands times between 10 to 20 times**  
**every time visted resources times increased the raised hands increased**

```python
# relation between visited resourses times & announcements view times regarding to the gender of student

plt.figure(figsize= (25, 5))

plt.subplot(2, 3, 1)
sns.regplot(data= df[df['gender'] == 'Male'], x= 'announcements_view', y= 'visited_resources')
plt.title('Relation Between announcements_view & visited_resources of Male student')

plt.subplot(2, 3, 2)
sns.histplot(data= df[df['gender'] == 'Male'], x= 'announcements_view', bins= 10)
plt.title('Distribution of announcements_view of Male student')

plt.subplot(2, 3, 3)
sns.histplot(data= df[df['gender'] == 'Male'], x= 'visited_resources', bins= 10)
plt.title('Distribution of visited_resources of Male student')

plt.subplot(2, 3, 4)
sns.regplot(data= df[df['gender'] == 'Female'], x= 'announcements_view', y= 'visited_resources')
plt.title('Relation Between announcements_view & visited_resources of Femal student')

plt.subplot(2, 3, 5)
sns.histplot(data= df[df['gender'] == 'Female'], x= 'announcements_view', bins= 10)
plt.title('Distribution of announcements_view of Femal student')

plt.subplot(2, 3, 6)
sns.histplot(data= df[df['gender'] == 'Female'], x= 'visited_resources', bins= 10)
plt.title('Distribution of visited_resources of Female student')
plt.tight_layout()
plt.show()
```

# Insights  
**max visited resources times from male between 80 to 90 times**  
**max visited resources times from female between 80 to 90 times**  
**max announcements view times from male is between 1 to 10 times**  
**max announcements view times from female is between 50 to 60  and between 20 to 30 times**

```python
# relation between visited resourses times & announcements view times regarding to the nationality of student

px.scatter(data_frame=df,\
           x='visited_resources',\
           y='announcements_view',\
           size='raised_hands',\
           color='nationality',\
           title='relation between visited resourses times & announcements view times regarding to the nationality of student')
```

# Insights  
**maximum of announcement view times records were by saudi arabia students with 100 raised hands times**  
**maximum of visited resources times records was by Kuwait & Jordan students with 70 & 85 raised hands times respectively**

```python
# relation between visited resourses times & announcements view times regarding to the gender of student & his nationality

px.scatter(data_frame=df,\
           x='visited_resources',\
           y='announcements_view',\
           size='raised_hands',\
           color='nationality',\
           facet_col='gender',\
           title='relation between visited resourses times & announcements view times regarding to the gender of student & his nationality')
```

# Insights  
**maximum of announcement view times records were by Kuwait male students with 72 raised hands times and 80 visited resources times**  
**maximum of visited resources times records were by Palestine male students with 90 raised hands times and 41 announcement view times**  
**maximum of announcement view times records were by Saudi Arabia female students with 100 raised hands times and 91 visited resources times**  
**maximum of visited resources times records were by Jordan female students with 85 raised hands times and 42 announcement view times**

```python
# count of students for each gender

px.bar(df.groupby('gender').agg({'gender': 'count'}),\
       color=df.groupby('gender').agg({'gender': 'count'}).index,\
       text_auto=True,\
       title='count of students for each gender',\
       height=400,\
       width=800)
```

# Insights

```python
# count of students for each gender regarding to nationality

px.bar(df.pivot_table(index='nationality', columns='gender', values='place_of_birth', aggfunc='count'),\
       barmode='group',\
       title='count of students for each gender regarding to nationality',\
       text_auto=True,\
       height=400)
```
